```r
library(rotl)
```

```r
names <- c(
    #'Homo sapiens',        # human
    'Mus musculus',        # mouse
    'Gallus gallus domesticus',  # chicken
    'Sus scrofa domesticus',     # pig
    'Solanum lycopersicum',  # tomato
    'Oryza sativa',          # rice
    'Bos taurus',            # cattle
    'Triticum aestivum',     # wheat
    'Solanum tuberosum',     # potato
    'Zea mays',              # maize
    'Ovis aries',            # sheep
    'Canis lupus familiaris'
)
```

```r
taxon_search <- tnrs_match_names(names = names, context_name = "All life")
```

```r
ott_in_tree <- ott_id(taxon_search)[is_in_tree(ott_id(taxon_search))]
tr <- tol_induced_subtree(ott_ids = ott_in_tree)
```

```r
plot(tr,direction="up",)
```

```r
pdf("figures/xgboost/species_tree.pdf", width = 3, height = 3)
plot(tr,direction = "up")
dev.off()
```
