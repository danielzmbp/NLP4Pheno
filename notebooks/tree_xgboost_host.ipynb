```r
library(rotl)
```

```r
names <- c(
	"Arabidopsis thaliana",
	"solanum tuberosum",
	"zea mays",
	"glycine max",
	"triticum aestivum",
	"Oryza sativa",
	"Solanum lycopersicum",
	"medicago"
)
```

```r
taxon_search <- tnrs_match_names(names = names, context_name = "All life")
```

```r
ott_in_tree <- ott_id(taxon_search)[is_in_tree(ott_id(taxon_search))]
tr <- tol_induced_subtree(ott_ids = ott_in_tree)
```

```r
plot(tr,direction="up",)
```

```r
pdf("figures/xgboost/species_tree.pdf", width = 3, height = 3)
plot(tr,direction = "up")
dev.off()
```
