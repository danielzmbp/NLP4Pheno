```python
from glob import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
assemblies = glob("/home/gomez/gomez/assemblies_linkbert/*/*/*.faa")
```

```python
len(assemblies)
```

```python
import pandas as pd
from collections import Counter

# Count the number of assemblies for each strain
assembly_counts = Counter([assembly.split('/')[-3] for assembly in assemblies])

# Create a dataframe from the assembly counts
df = pd.DataFrame({'Strain': list(assembly_counts.keys()), 'Assembly Count': list(assembly_counts.values())})

# Print the dataframe
print(df)
```

```python
df[df["Assembly Count"]<20].groupby("Assembly Count").count().plot(kind='line', figsize=(16, 6))
plt.axhline(y=10, color='r', linestyle='-')
```
