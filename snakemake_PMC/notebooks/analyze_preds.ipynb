```python
from glob import glob
import pandas as pd
```

```python
dfs = []
for f in glob("../*_filelist.tsv"):
	df = pd.read_csv(f, sep="\t")
	dfs.append(df)
dfs = pd.concat(dfs)
```

```python
dfs
```

```python
s = dfs["Article Citation"].str.split(".", expand=True)[0].value_counts()
```

```python
chunk_size = 1500
for i in range(0, len(s.index), chunk_size):
	chunk = s.index[i:i+chunk_size]
	print(chunk.tolist(),"\n")
```
